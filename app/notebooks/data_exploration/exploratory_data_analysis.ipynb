## Functions

```python
# Function that creates a dataframe for the root object
def create_root_df(df):
    return df[['_id', 'imei', 'dts', 'dtd', 'buff']]

# Functiont that creates a dataframe for the event object that is nested in the root object, this datafame contains all the variables that start with "event." and sorts the columns in alphabetical order
def create_event_df(df):
    return df.filter(regex='^event\.', axis=1).sort_index(axis=1)

# Function that  creates a dataframe for the device object that is nested in the root object, this dataframe contains all the variables that start with "device."
def create_device_df(df):
    return df.filter(regex='^device\.', axis=1).sort_index(axis=1)

# Function that creates a dataframe for the tracker object that is nested in the root object, this dataframe contains all the variables that start with "tracker."
def create_tracker_df(df):
    return df.filter(regex='^tracker\.', axis=1).sort_index(axis=1)

# Function that creates a dataframe for the crash object that is nested in the root object, this dataframe contains all the variables that start with "crash."
def create_crash_df(df):  
    return df.filter(regex='^crash\.', axis=1).sort_index(axis=1)
```

```python
# Function that shows the number of missing values in each column for a dataframe
def show_missing_values(df):
    return df.isnull().sum()
```

```python
root_object = create_root_df(df)
event_object = create_event_df(df)
device_object = create_device_df(df)
tracker_object = create_tracker_df(df)
crash_object = create_crash_df(df)
```

```python
# show missinhg values for each object, but sort the missing values in descending order
missing_values_root = show_missing_values(root_object).sort_values(ascending=False)
missing_values_event = show_missing_values(event_object).sort_values(ascending=False)
missing_values_device = show_missing_values(device_object).sort_values(ascending=False)
missing_values_tracker = show_missing_values(tracker_object).sort_values(ascending=False)
missing_values_crash = show_missing_values(crash_object).sort_values(ascending=False)
```

```python
# show all the variables above under each other with a line of space between them
print(missing_values_root, '\n')
print(missing_values_event, '\n')
print(missing_values_device, '\n')
print(missing_values_tracker, '\n')
print(missing_values_crash, '\n')
```
